```python
URL = "https://inshorts.com/en/read/sports"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html.parser')
news = soup.find_all("div", class_="content")
for i in news:
    title=i.find("span", itemprop="headline")
    print(title.text)
    #article=i.find("div", class_="news-card-content news-right-box")
    #print(article.text)
    #print()
    #row=[title.text,article.text,'Sports']
    #rows.append(row)
```

```python
URL = "https://www.indiatvnews.com/sports"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html.parser')
news = soup.find_all("li", class_="p_news")
for i in news:
    content=i.find("div", class_="content")
    #print(content)
    title=content.find("h3", class_="title")
    #print(title.text)
    article=content.find("p", class_="dic")
    #print(article.text)
    #print()
    row=[title.text,article.text]
    rows.append(row)
```

```python
URL = "https://www.india.com/sports/"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html.parser')
#print(soup)
news = soup.find_all("article", class_="repeat-box")
for i in news:
    figcaption=i.find("figcaption",class_="text")
    if figcaption!=None:
        titles=figcaption.find("h2")
        if titles!=None:
            title=(titles.text)
        para=figcaption.findChildren("p")
        if len(para)!=0:
            article=para[0].text
            #print(article
        row=[title,article,"Sports"]
        rows.append(row)
```

```python
with open('sports.csv', 'w', newline='',encoding='utf-8') as file:
     writer = csv.writer(file)
     writer.writerows(rows)
```
